```python
from bs4 import BeautifulSoup
```

```python
with open("400 Python Job Vacancies in Ahmedabad, Gujarat ｜ Indeed (5_29_2025 2：00：47 PM).html") as f:
    html=f.read()
```

```python
soup=BeautifulSoup(html,"html.parser")
z=soup.find_all("div",{"class":"job_seen_beacon"})
```

```python
c=[]
c1=[]
city=[]
s=[]
for i in z:
    cm=i.find("h2",{"class":"jobTitle css-1psdjh5 eu4oa1w0"})
    c.append(cm.text)
    cm1=i.find("span",{"class":"css-1h7lukg eu4oa1w0"})
    c1.append(cm1.text)
    city1=i.find("div",{"class":"css-1restlb eu4oa1w0"}).text.split(",")[0]
    city.append(city1)
    s1=i.find("div",{"class":"css-1restlb eu4oa1w0"}).text.split(",")[1]
    s.append(s1)
```
