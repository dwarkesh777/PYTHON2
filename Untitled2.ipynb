```python
from bs4  import BeautifulSoup
```

```python
with open("newflipkart.html","r",encoding="utf-8") as f:
    html=f.read()
```

```python
soup=BeautifulSoup(html,"html.parser")
```

```python
z=soup.find_all("div",class_="_75nlfW")
```

```python
name=[]
ram=[]
color=[]
star=[]
review=[]
old=[]
new=[]
dis=[]
exc=[]
for i in z:
    n=i.find("div",class_="KzDlHZ").text
    name.append(n.split("(")[0].strip())
    ram.append(n.split("(")[1].split(",")[1][0:-1].strip()+" rom")
    color.append(n.split("(")[1].split(",")[0].strip())
    s=i.find("div",class_="XQDdHH")
    star.append(s.text.strip())
    r=i.find("span",class_="Wphh3N").text.split("&")[1].split(" ")[0].strip()
    review.append(r)
    o=i.find("div",class_="yRaY8j ZYYwLA").text.split("₹")[1].replace(",","").strip()
    old.append(int(o))
    n=i.find("div",class_="Nx9bqj _4b5DiR").text.split("₹")[1].replace(",","").strip()
    new.append(int(n))
    d=i.find("div",class_="UkUFwK").text.split(" ")[0].strip()
    dis.append(d)
    e=i.find("div",class_="col col-5-12 BfVC2z").text
    if "DealUpto" in e:
        exc.append(e.split("DealUpto ₹")[1].split(" ")[0].strip().replace(",",""))
    else:
        exc.append("Nan")
    
```

```python
import pandas as pd
```

```python
data={"name":name,"ram":ram,"color":color,"star":star,"review":review,"old":old,"new":new,"dis":dis,"exc":exc}
```

```python
df=pd.DataFrame(data)
```
