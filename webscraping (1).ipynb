# find_all

```python
l1=soup.find_all("div",{"class":"student"},limit=2)
```

```python
name=[]
email=[]
dept=[]
social=[]
link=[]
ext=[]
img=[]
for i in l1:
    n=i.find("h2").text
    name.append(n)
    e=i.find("span").text.split("@")[0]
    email.append(e)
    x=i.find("span").text[-10:]
    ext.append(x)
    d=i.find("p").text
    dept.append(d)
    s=i.find("a").text.strip()
    social.append(s)
    l=i.find("a")
    link.append(l["href"])
    im=i.find("img")
    img.append(im["src"])
```

```python
data={"name":name,"email":email,"ext":ext
      ,"dep":dept,"SOCIAL":social,"link":link,"img":img}
```

```python
data
```

```python
df=psd.DataFrame(data)
```
