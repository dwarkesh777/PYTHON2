```python
lat=25.276987
lon=55.296249
key="e70a7528249e1c3633e76ba75a6897bd"
url=f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={key}"
data=requests.get(url)
data.status_code
```

```python
final=data.json()
```

```python
import json
new=json.dumps(final,indent=5)
```

```python
print(new)
```

```python
d=[]
t=[]
tk=[]
tc=[]
ws=[]
tf=[]
for i in final["list"]:
    d.append(i["dt_txt"].split(" ")[0])
    t.append(i["dt_txt"].split(" ")[1])
    tk.append(i["main"]["temp"])
    tc.append(i["main"]["temp"]-272.5)
    ws.append(i["wind"]["speed"])
    tf.append((i["main"]["temp"]-272.5)*9/5+32)
```

```python
data={"date":d,"time":t,"tk":tk,"tc":tc,"tf":tf,"wind":ws}
```

```python
df=pd.DataFrame(data)
df
```

```python
x=df[df["time"]=="12:00:00"]
```
